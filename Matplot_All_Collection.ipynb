## Matplotlib, pyplot and pylab: how are they related?
Matplotlib is the whole package and :mod:`matplotlib.pyplot` is a module in
Matplotlib.

For functions in the pyplot module, there is always a "current" figure and
axes (which is created automatically on request).  For example, in the
following example, the first call to ``plt.plot`` creates the axes, then
subsequent calls to ``plt.plot`` add additional lines on the same axes, and
``plt.xlabel``, ``plt.ylabel``, ``plt.title`` and ``plt.legend`` set the
axes labels and title and add a legend.

```python
x=np.linspace(0,2,10) # create elements  n = 100 from 0 to 10
plt.plot(x,x,label= "Linear")
plt.plot(x,x**2,label = "Square")
plt.plot(x*3,  label = "Cubic")

plt.xlabel("X label")
plt.ylabel("Y column")

plt.title("This is first graph")
plt.legend()
plt.show()
```

### Stacked Bar Graph

```python
import numpy as np
import matplotlib.pyplot as plt

N = 5
menMeans = (20,35,30,35,27)
womenMeans = (25,32,34,20,25)
menStd = (2,3,4,1,2)
womenStd = (3,5,4,1,2)
ind = np.arange(N)
width = 0.3  #widhth of bar can be changed as per requirements
p1 = plt.bar(ind , menMeans , width,yerr=menStd, label= "MenMeans")
p2 = plt.bar(ind, womenMeans , width , bottom=menMeans, yerr=womenMeans, label = "Women")

plt.ylabel("Scores")
plt.title("Title of this Graf")
plt.xticks(ind, ("Seoul", "Kathmandu", "Newyork", "Pusan", "London"))
plt.yticks([10,20,30,40,50,60,70,80])
plt.legend()
plt.show()
```

## Grouped bar chart with labels

```python
import matplotlib.pyplot as plt
import numpy as np

labels= ["Seoul", "Kathmandu", "Newyork", "Pusan", "London"]#any descrete values
men_means = [20,34,30,35,27]
women_means = [25,32,34,20,25]

x = np.arange(len(labels)) #label Location
width = 0.40
fig, ax = plt.subplots()
rect1 = ax.bar(x-width/2,men_means , width, label = "Men") #x 
rect2 = ax.bar (x+width/2,women_means,width , label =" Women")

plt.xticks()
plt.legend()


plt.show()
```


## Broken Barh

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.broken_barh([(110, 30), (150, 10)], (10, 9), facecolors='tab:blue')
ax.broken_barh([(10, 50), (100, 20), (130, 10)], (20, 9),
               facecolors=('tab:orange', 'tab:green', 'tab:red'))
ax.set_ylim(5, 35)
ax.set_xlim(0, 200)
ax.set_xlabel('seconds since start')
ax.set_yticks([15, 25])
ax.set_yticklabels(['Bill', 'Jim'])
ax.grid(True)
ax.annotate('race interrupted', (61, 25),
            xytext=(0.8, 0.9), textcoords='axes fraction',
            arrowprops=dict(facecolor='black', shrink=0.05),
            fontsize=16,
            horizontalalignment='right', verticalalignment='top')

plt.show()
```

## Text WaterMark

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(2000)

fig,ax = plt.subplots()
ax.plot(np.random.rand(20),"-o",ms=20,lw=2, alpha=0.7,mfc=ax.grid())

fig.text(0.95,0.5,"Krishna ",
        fontsize = 50, color="r",
        ha= "right",va="top", alpha = 15)
plt.show()
```
